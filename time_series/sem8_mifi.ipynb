# LSTM 1 вариант с фичей как lag_MA7 , данные не отскейлены

```python
# Create sequences for training (input: seq_length values, output: next value)
def create_sequences(data, seq_length):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data.iloc[i:(i + seq_length)]['lag_MA7']
        y = data.iloc[i + seq_length]['Close']
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_length = 30  # Length of input sequence
X, y = create_sequences(df, seq_length)
```

```python
# Convert to PyTorch tensors
# Reshape X to (batch_size, seq_length, input_size)
trainX = torch.tensor(X[:, :, None], dtype=torch.float32)
trainY = torch.tensor(y[:, None], dtype=torch.float32)
```

```python
len(trainX)
```

```python
class LSTMModel(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout=0.2):
        super(LSTMModel, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # начальные состояния (h0, c0)
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size)

        # пропускаем через LSTM
        out, _ = self.lstm(x, (h0, c0))
        out = self.fc(out[:, -1, :])  # берем последнее состояние
        return out

# параметры модели
input_size = trainX.shape[2]
hidden_size = 64
num_layers = 2
output_size = 1

model = LSTMModel(input_size, hidden_size, num_layers, output_size)

# =======================
# 4. Обучение модели
# =======================
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

n_epochs = 150
for epoch in range(n_epochs):
    model.train()
    optimizer.zero_grad()
    output = model(trainX)
    loss = criterion(output.flatten(), trainY)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{n_epochs}], Loss: {loss.item():.6f}")

# =======================
# 5. Прогноз
# =======================
model.eval()
with torch.no_grad():
    y_pred = model(trainX).flatten()
```

```python
def plot_values(df, X, y_pred):
    plt.plot(df.index, df['Close'], label='Фактические значения', color='green', linewidth=2)
    plt.plot(df.index[-len(X):], y_pred, label='Предсказанные значения', color='red', linestyle='--', linewidth=2)
    plt.title('Фактические vs Предсказанные значения')
    plt.xlabel('Дата')
    plt.ylabel('Цена закрытия')
    plt.legend()
    plt.grid(True, alpha=0.3)
    plt.xticks(rotation=45);
```

```python
plot_values(df, X, y_pred)
```

```python
class SimpleMLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleMLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size) # First fully connected layer
        self.relu = nn.ReLU()                       # Activation function
        self.fc2 = nn.Linear(hidden_size, hidden_size) # First fully connected layer
        self.relu2 = nn.ReLU()                       # Activation function
        self.fc3 = nn.Linear(hidden_size, output_size) # Second fully connected layer

    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.fc3(out)
        return out

# Example usage:
input_size = 180
hidden_size = 20
output_size = 1

# Instantiate the model
model = SimpleMLP(input_size, hidden_size, output_size)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

n_epochs = 5550
for epoch in range(n_epochs):
    model.train()
    optimizer.zero_grad()
    output = model(trainX)
    loss = criterion(output.flatten(), trainY)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 550 == 0:
        print(f"Epoch [{epoch+1}/{n_epochs}], Loss: {loss.item():.6f}")

# =======================
# 5. Прогноз
# =======================
model.eval()
with torch.no_grad():
    y_pred = model(trainX).flatten()
```
